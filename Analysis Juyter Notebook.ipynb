```python
def parse_hours(hours_str):
    # Check for NaN (missing data)
    if pd.isna(hours_str):
        return pd.Series([None, None])
    
    # Remove the leading 'hours': part of the string and strip leading/trailing spaces
    hours_str = hours_str.replace("'hours':", "").strip()
    
    # Handle the 'Closed' entries
    if hours_str.lower() == "'closed'":
        return pd.Series([None, None])

    # Now split the string by 'to'
    try:
        start_time, end_time = hours_str.split(' to ')
        return pd.Series([start_time.strip(), end_time.strip()])
    except ValueError:
        # Handle any unexpected formats by returning None
        return pd.Series([None, None])

# Apply the function to the 'Hours' column to create two new columns
Tuesday[['Start Time', 'End Time']] = Tuesday['Hours'].apply(parse_hours)

Tuesday
```
